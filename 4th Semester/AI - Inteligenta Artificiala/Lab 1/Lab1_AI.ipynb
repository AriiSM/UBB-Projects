```python
# 1) Să se determine ultimul (din punct de vedere alfabetic) cuvânt care poate apărea într-un text care conține mai
# multe cuvinte separate prin ” ” (spațiu). De ex. ultimul (dpdv alfabetic) cuvânt din ”Ana are mere rosii si galbene” este cuvântul "si"

#functie
def Pb1Sor(l):
  txt = l.split()
  lista = sorted(txt, reverse=True)
  return lista[0]

#Bubble Sort
#O(n^2)
def Pb1Sor2(l):
  txt = l.split()
  for i in range(len(txt)):
        for j in range(0, len(txt)-i-1):
            if txt[j] < txt[j+1]:
                txt[j], txt[j+1] = txt[j+1], txt[j]
  return txt[0]
#test
def Test1Sor(cerinta):
  print("Testare start")
  assert(cerinta("Ana are mere rosii si galbene") == 'si')
  assert(cerinta("a b c") == 'c')
  print("Succes")

#O(n)
def ultimul_cuvant_alfabetic(text):
    # Separă textul în cuvinte
    cuvinte = text.split()
    # Sortează cuvintele în ordine alfabetică inversă
    cuvinte_sortate = sorted(cuvinte, reverse=True)
    # Returnează primul cuvânt din lista sortată (adică ultimul din punct de vedere alfabetic)
    return cuvinte_sortate[0]



#apeluri
Test1Sor(Pb1Sor)
Test1Sor(Pb1Sor2)
Test1Sor(ultimul_cuvant_alfabetic)

l = "Ana are mere rosii si galbene"
print("Metoda1: ",Pb1Sor(l))
print("Metoda2: ",Pb1Sor2(l))



#Copilot
text = "Ana are mere rosii si galbene"
ultimul = ultimul_cuvant_alfabetic(text)
print("Copilot: Ultimul cuvânt alfabetic este:", ultimul)
```

```python
import math
from math import hypot
#2)
#Să se determine distanța Euclideană între două locații identificate prin perechi de numere. De ex. distanța între (1,5) și (4,1) este 5.0

# x1,x2 coordonatele lui X
# y1,y2 coordonatele lui Y
#output: distanta Eucliiana calculata cu formula
#Complexitate O(1)
def Pb2Euc1(x1,x2,y1,y2):
  dx = int(x2)-int(x1)
  dy = int(y2)-int(y1)
  rez = math.sqrt(pow(dx,2) + pow(dy,2))
  return rez

#The math.hypot() method returns the Euclidean norm.
#Complexitate O(1)
def Pb2Euc2(x1,x2,y1,y2):
    dx = x2 - x1
    dy = y2 - y1
    distance = hypot(dx, dy)
    return distance

#teste
def TestPb2(cerinta):
    print("Test Start")
    assert(cerinta(1,5,4,1) == 5.0)
    assert(cerinta(0,0,0,0) == 0.0)
    print("Succes")




#Complexitate O(1)
def Pb2Copilot(x1,x2,y1,y2):

    dx = x2 - x1
    dy = y2 - y1

    # Calculează distanța Euclideană folosind teorema lui Pitagora
    distance = math.sqrt(dx**2 + dy**2)
    return distance


TestPb2(Pb2Euc1)
TestPb2(Pb2Euc2)
TestPb2(Pb2Copilot)

print("Coordonatele lui X")
x1 = 1
x2 = 5
print ("Coordonatele lui Y")
y1 = 4
y2 = 1
print("1)Distanta Euclidiana1:",Pb2Euc1(x1,x2,y1,y2))
print("2)Distanta Euclidiana1:",Pb2Euc2(x1,x2,y1,y2))




#Copilot
point1 = (1, 5)
point2 = (4, 1)
# Calculează distanța Euclideană
distance = Pb2Copilot(1, 5, 4, 1)
print("Copilot: Distanța Euclideană între", point1, "și", point2, "este:", distance)
```

```python
from collections import defaultdict
#  4) Să se determine cuvintele unui text care apar exact o singură dată în acel text.
#  De ex. cuvintele care apar o singură dată în ”ana are ana are mere rosii ana" sunt: 'mere' și 'rosii'.

#Vector de frecventa
#complexitate O(n), n-nr cuvinte
def Pb4Cuv(IP):
  #facem split pe cuvinte (default pe spatiu)
  text = IP.split()
  #definim un vector de frecventa
  frecventa = {}
  #parcurgem fiecare cuvant din propozitie
  for cuv in text:
      #verificam daca cuv a mai aparut in vector de frecventa
      if cuv in frecventa:
          #si incrementam valoarea
          frecventa[cuv] += 1
      else:
          #altfel adaugam noul cuvant si ii dam frecventa de 1
          frecventa[cuv] = 1
  rezultat=[]
  #stocam cuvintele care apar o singura data intr-un vector
  for key, value in frecventa.items():
    if value == 1:
      rezultat.append(key)
  return rezultat

#cu dictionar
#O(n)
def Pb4Cuv2(IP):
  text = IP.split()
  lista = sorted(text)
  rezultat=[]
  #dictionar
  aparitii = defaultdict(list)
  for cuvant in lista:
    aparitii[cuvant].append(cuvant)

  rezultat = []
  for cuvant, lista_aparitii in aparitii.items():
    if len(lista_aparitii) == 1:
      rezultat.append(lista_aparitii[0])

  return rezultat

def TestPb4Cuv(cerinta):
  print("Test Start")
  assert(len(cerinta('ana are ana are mere rosii ana'))==2)
  assert(len(cerinta('ana are ana are mere mere rosii rosii ana'))==0)
  assert(len(cerinta('ana are mere'))==3)
  print("Succes!")



#complexitate O(n), n-nr cuvinte
def Pb4Copilot(text):
  # Impartim textul in cuvinte
  cuvinte = text.split()

  # Cream un dictionar pentru a stoca frecventa cuvintelor
  frecventa_cuvinte = {}

  # Numaram frecventa fiecarui cuvant
  for cuvant in cuvinte:
      cuvant = cuvant.lower()  # Convertim la litere mici
      if cuvant in frecventa_cuvinte:
          frecventa_cuvinte[cuvant] += 1
      else:
          frecventa_cuvinte[cuvant] = 1
  # Gasim cuvintele care apar exact o singura data
  cuvinte_unice = [cuvant for cuvant, frecventa in frecventa_cuvinte.items() if frecventa == 1]
  return cuvinte_unice



#teste cod + GBT
TestPb4Cuv(Pb4Cuv);
TestPb4Cuv(Pb4Cuv2);
TestPb4Cuv(Pb4Copilot);

IP = "ana are ana are mere rosii ana"
txt = Pb4Cuv(IP)
txt1 = Pb4Cuv2(IP)

print("1)Cuvintele care apar o singura data:")
for i in txt:
  print(i)
print("2) Cuv care apar o sg data", txt1)



#copilot
print(f"Copilot: Cuvintele unice din text sunt: {', '.join(Pb4Copilot(IP))}")
```

```python
#5) Pentru un șir cu n elemente care conține valori din mulțimea {1, 2, ..., n - 1}
# astfel încât o singură valoare se repetă de două ori, să se identifice acea valoare care se repetă.
# De ex. în șirul [1,2,3,4,2] valoarea 2 apare de două ori.

#creez un vector de frecventa si cand acesta creste la 2 ii afisez cheia
#O(n)
def Pb5Rep(l):
  frecventa = {}
  nr = 0
  for i in l:
    if i != " ":
      if i in frecventa:
          frecventa[i] += 1
      else:
          frecventa[i] = 1
      if frecventa[i] == 2:
        nr = i
        break
  return nr

#Parcurgere lista de numere, daca nu exista in set la adaug, daca exista in set afisez acel numar
#O(n)
def Pb5Rep2(l):
  numere = set()
  for i in l:
    if i !=" ":
      if i in numere:
        return i
      else:
        numere.add(i)
  return 0

def TestPb5(cerinta):
    print("Test Start")
    assert cerinta([1, 2, 3, 4, 2]) == 2
    print("Succes")



#Complexitate O(1)
def Pb5Copilot(nums):
    n = len(nums)
    # Calculează suma elementelor din șir
    sum_of_elements = sum(nums)
    # Calculează suma primelor n-1 numere naturale
    sum_of_natural_numbers = (n - 1) * n // 2

    # Diferența dintre sumele este valoarea care se repetă
    duplicate_value = sum_of_elements - sum_of_natural_numbers

    return duplicate_value



TestPb5(Pb5Rep)
TestPb5(Pb5Rep2)
TestPb5(Pb5Copilot)

l = "1 2 3"
print("1)Nr care apare de 2 ori este:",Pb5Rep(l))
print("2)Nr care apare de 2 ori este:",Pb5Rep2(l))



# Exemplu de utilizare
arr = [1, 2, 3, 4, 2]
duplicate = Pb5Copilot(arr)
print("Copilot: Valoarea care se repetă în șirul", arr, "este:", duplicate)
```

```python
from collections import defaultdict
# 6)Pentru un șir cu n numere întregi care conține și duplicate, să se determine elementul majoritar
#  (care apare de mai mult de n / 2 ori). De ex. 2 este elementul majoritar în șirul [2,8,7,2,2,5,2,3,1,2,2].

#follsing vector de frecventa
#O(n)
def Pb6Max(l):
  frecventa = {}
  nr = 0
  for i in l:
    if i in frecventa:
        frecventa[i] += 1
    else:
        frecventa[i] = 1
  lung = int(len(l) / 2)
  rezultat = set()
  for key, value in frecventa.items():
    if value >= lung:
      rezultat.add(key)
  return rezultat


#folisim dictionar
#O(n)
def Pb6Max2(l):
  lista = sorted(l)
  lung = int(len(l) / 2)
  rezultat=[]
  aparitii = defaultdict(list)
  for nr in lista:
    aparitii[nr].append(nr)

  for nr, ap in aparitii.items():
    if len(ap) >= lung:
      rezultat.append(ap[0])

  return rezultat

def TestPb6(cerinta):
  print("Testare start")
  if cerinta == "Pb6Copilot":
    assert(cerinta([2, 8, 7, 2, 2, 5, 2, 3, 1, 2, 2])==2)
  else:
    assert(cerinta([1,1,2,2,3]) == {1, 2})
    assert(cerinta([2,8,7,2,2,5,2,3,1,2,2]) == {2})
  print("Succes")



# O(n)
def Pb6Copilot(nums):
    candidate = None
    count = 0

    for num in nums:
        if count == 0:
            candidate = num
            count = 1
        elif candidate == num:
            count += 1
        else:
            count -= 1

    # Verificăm dacă candidatul este de fapt element majoritar
    count = 0
    for num in nums:
        if num == candidate:
            count += 1
    if count > len(nums) / 2:
        return candidate
    else:
        return None


TestPb6(Pb6Max)
# TestPb6(Pb6Copilot)

l = [1,1,2,2,3]
print("Raspuns pentru: [2,8,7,2,2,5,2,3,1,2,2]: ", Pb6Max(l))
print("Raspuns pentru: [2,8,7,2,2,5,2,3,1,2,2]: ", Pb6Max2(l))



arr = [2, 8, 7, 2, 2, 5, 2, 3, 1, 2, 2]
majority_element = Pb6Copilot(arr)
if majority_element is not None:
    print("Copilot: Elementul majoritar în șirul", arr, "este:", majority_element)
else:
    print("Copilot: Nu există element majoritar în șirul", arr)
```
