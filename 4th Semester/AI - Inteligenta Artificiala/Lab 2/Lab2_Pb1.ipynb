```python
#e)in cazul proprietatilor nenumerice, cate valori posibile are fiecare astfel de proprietate
coloana = employees['First Name'].dropna().unique()
print("First Name:  ", len(coloana))

coloana = employees['Gender'].dropna().unique()
print("Gender:  ",len(coloana))

coloana = employees['Senior Management'].dropna().unique()
print("Senior Management:  ",len(coloana))

coloana = employees['Team'].dropna().unique()
print("Team:  " ,len(coloana))
```

```python
#f)daca sunt valori lipsa si cum se poate rezolva aceasta problema
total_completate = 0
total_necompletate = 0
#Parcurs coloanele
for coloana in employees.columns:
    # Verifică dacă câmpurile sunt completate sau nu în coloana curentă
    campuri_necompletate = employees[employees[coloana].isnull()]
    campuri_completate = employees[employees[coloana].notnull()]
    total_completate += len(campuri_completate)
    total_necompletate += len(campuri_necompletate)

print("Valori Completate: " + str(total_completate))
print("Valori Necompletate: " + str(total_necompletate))

if total_necompletate != 0:
    print("Se pot sterge toate inregistrarile care au zone necompletate!")

fisier_employees_complet = employees.dropna()
print("Numarul de inregistrari dupa stergerea datelor incomplete este de: ", len(fisier_employees_complet))
```

```python
import matplotlib.pyplot as plt
#1.b. Sa se vizualizeze:
#a)distributia salariilor acestor angajati pe categorii de salar
plt.hist(fisier_employees_complet['Salary'], bins = 50,edgecolor='black')  
plt.xlabel('Valoare')
plt.ylabel('Frecventa')
plt.title('Distributia angajati')
plt.show()
```

```python
#b)distributia salariilor acestor angajati pe categorii de salar si echipa din care fac parte
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict

# make data:
salarii = fisier_employees_complet['Salary']
team = fisier_employees_complet['Team'].unique()

salarii_pe_echipe = defaultdict(list)
for index, row in fisier_employees_complet.iterrows():
    if row['Team'] in team:
        salarii_pe_echipe[row['Team']].append(row['Salary'])
    
fig, ax = plt.subplots()
ax.boxplot(salarii_pe_echipe.values(),widths=0.5, patch_artist=True,
                showmeans=False, showfliers=True,
                medianprops={"color": "blue", "linewidth": 1.5},
                boxprops={"facecolor": "grey", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "black", "linewidth": 1.5},
                capprops={"color": "black", "linewidth": 1.5},
                flierprops=dict(marker='o', markersize=5, markerfacecolor='red', markeredgecolor='red'))
fig.set_size_inches(20, 7)
ax.set_xticklabels(salarii_pe_echipe.keys())
ax.set_xlabel('Echipa')
ax.set_ylabel('Salariu')
ax.set_title('Salarii pe echipă')
plt.show()
```
